```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data_labels = ['[9]', '[12]', '[47]', '[6]', '[7]', '[13] bigdata', '[13] small data', '[14]', '[8]', '[19]', '[20]', '[21]', '[22]', '[48]', 'Ours']
accuracy = [97.20, 86.00, 98.36, 90.40, 87.42, 96.80, 68.00, 96.89, 84.40, 0, 97.50, 0, 98.80, 99.96, 99.98]
f_score = [97.60, 81.00, 98.12, 92.80, 82.10, 96.80, 69.30, 0, 85.00, 95.78, 95.00, 93.30, 98.90, 99.96, 99.98]

x = np.arange(len(data_labels))  # Label locations
width = 0.4  # Bar width

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, accuracy, width, label='Accuracy', color='steelblue')
rects2 = ax.bar(x + width/2, f_score, width, label='F1-Score', color='salmon')

# Labels and Title
ax.set_xlabel("Methods")
ax.set_ylabel("Percentage")
ax.set_title("Comparison of Accuracy and F Score", pad=20)
ax.set_xticks(x)
ax.set_xticklabels(data_labels, rotation=90, ha="right")
# ax.legend()
ax.legend(bbox_to_anchor=(0.1, -0.15), loc='upper center', borderaxespad=0.)
ax.set_ylim(0, 102)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.2f}'))

# Add grid lines
ax.yaxis.grid(True, linestyle='-', linewidth=0.6, color='grey')

# Show plot
plt.tight_layout()
plt.show()
```
